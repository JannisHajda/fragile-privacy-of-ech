```python
import matplotlib.pyplot as plt
import pandas as pd

# Data: Parsing manually from the provided text
data = {
    "Date": [
        "2024-08-08_18-18-05",
        "2024-08-19_19-04-54",
        "2024-08-21_13-59-28",
        "2024-09-05_01-33-25",
        "2024-09-06_01-04-14",
        "2024-09-08_14-44-57",
        "2024-09-11_10-14-23",
        "2024-09-15_21-43-55",
        "2024-10-13_02-50-30",
    ],
    "ECH Configurations": [
        30,
        15301,
        67979,
        297137,
        592346,
        592006,
        1029773,
        1177134,
        1165138,
    ],
    "Distinct ECH Configurations": [16, 25, 24, 26, 25, 25, 27, 24, 45],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the 'Date' column to datetime for better plotting
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d_%H-%M-%S")

# Set Date as the index for easier plotting
df.set_index("Date", inplace=True)

# Set up the figure with larger dimensions for better readability
fig, ax = plt.subplots(figsize=(14, 8))

# Define bar width and spacing
width = 0.45  # Slightly thicker bars
spacing = 7  # Space between the result bars

# Plot ECH Configurations and Distinct ECH Configurations as horizontal bars
ax.barh(
    df.index - pd.Timedelta(hours=spacing),
    df["ECH Configurations"],
    height=width,
    label="ECH Configurations",
    color="#1f77b4",
    alpha=0.85,
)
ax.barh(
    df.index + pd.Timedelta(hours=spacing),
    df["Distinct ECH Configurations"],
    height=width,
    label="Distinct ECH Configurations",
    color="#2ca02c",
    alpha=0.85,
)

# Add a horizontal line on the activation date (August 16, 2024)
activation_date = pd.to_datetime("2024-08-14")
ax.axhline(
    activation_date, color="orange", linestyle="--", linewidth=2, xmin=0, xmax=0.95
)  # label='Cloudflare ECH Activation',

# Adjust the annotation position closer to the plot, using specific coordinates
ax.annotate(
    "Cloudflare Activation Start",
    xy=(0.77, 0.15),
    xycoords="axes fraction",
    color="orange",
    fontsize=12,
    ha="left",
    va="center",
)
ax.annotate(
    "2024-08-14",
    xy=(-0.072, 0.12),
    xycoords="axes fraction",
    color="orange",
    fontsize=10,
    ha="left",
    va="center",
)

# Add a horizontal line on the activation date (August 16, 2024)
activation_date = pd.to_datetime("2024-09-16")
ax.axhline(
    activation_date, color="orange", linestyle="--", linewidth=2, xmin=0, xmax=0.95
)  # label='Cloudflare ECH Activation',

# Adjust the annotation position closer to the plot, using specific coordinates
ax.annotate(
    "Cloudflare Activation End",
    xy=(0.77, 0.6),
    xycoords="axes fraction",
    color="orange",
    fontsize=12,
    ha="left",
    va="center",
)
ax.annotate(
    "2024-09-16",
    xy=(-0.072, 0.59),
    xycoords="axes fraction",
    color="orange",
    fontsize=10,
    ha="left",
    va="center",
)

# Use logarithmic scale on x-axis
ax.set_xscale("log")

# Hide axis titles
ax.set_xlabel("")
ax.set_ylabel("")

# Set y-ticks to correspond to each date
ax.set_yticks(df.index)
ax.set_yticklabels(df.index.strftime("%Y-%m-%d"), fontsize=10)

# Set tick parameters for x-axis
ax.tick_params(axis="x", which="major", labelsize=12)
ax.tick_params(axis="x", which="minor", labelsize=10)

# Drop ax Spines
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Add annotations for ECH Configurations below the bars
for i, v in enumerate(df["ECH Configurations"]):
    ax.text(
        v,
        df.index[i] - pd.Timedelta(hours=spacing),
        f"{v}",
        ha="left",
        va="center",
        fontsize=10,
        fontweight="bold",
        color="#1f77b4",
    )

# Add annotations for Distinct ECH Configurations below the bars
for i, v in enumerate(df["Distinct ECH Configurations"]):
    ax.text(
        v,
        df.index[i] + pd.Timedelta(hours=spacing),
        f"{v}",
        ha="left",
        va="center",
        fontsize=10,
        fontweight="bold",
        color="#2ca02c",
    )

# Add legend, adjusting its position to the top right
ax.legend(loc="lower right", fontsize=12)

# Tight layout for proper spacing
plt.tight_layout()

# Show plot
plt.show()
```
