```python
import os


def count_lines_in_files(folder_path):
    results = {}
    for filename in os.listdir(folder_path):
        filepath = os.path.join(folder_path, filename)
        if os.path.isfile(filepath) and filename.lower().endswith((".csv", ".txt")):
            try:
                with open(filepath, encoding="utf-8") as f:
                    line_count = sum(1 for line in f)
                results[filename] = line_count
            except UnicodeDecodeError:
                results[filename] = "Encoding error"
            except OSError:
                results[filename] = "File read error"
        elif os.path.isdir(filepath):
            results[filename] = "Directory"
        else:
            results[filename] = "Unsupported file type"
    return results


def format_with_dots(number):
    if isinstance(number, int):
        # Format number with dots for thousands separator
        return f"{number:,}".replace(",", ".")
    return number  # If not an int, return the error message or description


# Folder path containing CSV and TXT files
folder_path = "../10_sourcing_domains/"  # Replace with your folder path

# Check if folder path exists
if not os.path.exists(folder_path):
    print(f"Folder path {folder_path} does not exist.")
else:
    # Count lines in each CSV/TXT file in the folder
    line_counts = count_lines_in_files(folder_path)

    # Print results with formatting
    for file, count in line_counts.items():
        formatted_count = format_with_dots(count)
        print(f"{file}: {formatted_count}")
```
