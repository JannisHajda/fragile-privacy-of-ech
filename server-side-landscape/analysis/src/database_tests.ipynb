```python
# Import necessary libraries
import os

import mysql.connector
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()


# Database connection setup function
def connect_to_db():
    global db_connection, cursor
    try:
        db_connection = mysql.connector.connect(
            host=os.getenv("DB_HOST"),
            user=os.getenv("DB_USER"),
            password=os.getenv("DB_PASSWORD"),
            database=os.getenv("DB_NAME"),
        )
        cursor = db_connection.cursor(dictionary=True)
        print("Database connection established successfully.")
    except mysql.connector.Error as e:
        print(f"Error connecting to database: {e}")
        db_connection = None
        cursor = None


# Establish the database connection
connect_to_db()
```

```python
# Check if connection is established
if db_connection and cursor:
    try:
        # Query to find duplicate domains
        query = """
        SELECT domain, COUNT(*) AS count
        FROM DomainList
        GROUP BY domain
        HAVING count > 1
        """

        # Execute the query
        cursor.execute(query)

        # Fetch all the results
        duplicate_domains = cursor.fetchall()

        # Display the duplicate domains
        if duplicate_domains:
            print("Duplicate Domains:")
            for domain, count in duplicate_domains:
                print(f"{domain}: {count} occurrences")
        else:
            print("No duplicate domains found.")

    except mysql.connector.Error as e:
        print(f"Error executing query: {e}")
else:
    print("Database connection failed.")
```

```python
# Check if connection is established
if db_connection and cursor:
    try:
        # Query to find domains without any source
        query = """
        SELECT domain
        FROM DomainList
        WHERE id NOT IN (
            SELECT domain_list_id
            FROM DomainListDomainSource
        )
        """

        # Execute the query
        cursor.execute(query)

        # Fetch all the results
        domains_without_source = cursor.fetchall()

        # Count of domains without any source
        count_domains_without_source = len(domains_without_source)

        # Display the total count
        print(f"Total domains without any source: {count_domains_without_source}\n")

        # Display the list of domains if any
        if domains_without_source:
            print("Domains without any source:")
            for domain_tuple in domains_without_source:
                print(domain_tuple[0])  # Accessing the domain within the tuple
        else:
            print("No domains found without any source.")

    except mysql.connector.Error as e:
        print(f"Error executing query: {e}")
else:
    print("Database connection failed.")
```

```python
if cursor:
    cursor.close()
if db_connection:
    db_connection.close()
    print("Database connection closed.")
```
