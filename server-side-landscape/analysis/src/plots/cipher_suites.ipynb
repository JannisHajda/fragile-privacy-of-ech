## ECH Cipher Suites over Time


```python
from pathlib import Path

import altair as alt
import polars as pl

from src.utils.date import monthYear

DATA_DIR = Path("../../data/measurements")

df = pl.read_parquet(DATA_DIR / "ech_cipher_suites.parquet")
```

```python
base = alt.Chart(df)

# Adjusted Altair chart with swapped axes
comet_chart = (
    base.mark_circle(size=75)
    .encode(
        x=alt.X(
            "test_date:T",  # Move test_date to the Y-axis
            # title="Date",
            title=None,
            scale=alt.Scale(padding=10),
            axis=alt.Axis(format=monthYear),
        ),
        y=alt.Y(
            "cipher_suite:N",  # Move cipher_suite_label to the X-axis
            title="Cipher Suite",
        ),
        color=alt.Color(
            "cipher_suite_count:Q",
            title="Domain Count",
            scale=alt.Scale(type="log", scheme="turbo"),
        ),
    )
    .properties(
        height=250,  # Adjust height for more test_date entries
    )
    .configure_legend(
        orient="none",
        direction="horizontal",
        legendX=-100,
        legendY=285,
        gradientLength=175,
    )
    .configure_axisY(labelExpr="split(datum.label, ' with ')")
)

# Show the updated chart
comet_chart.save("../../plots/hpke_cipher_suite.pdf")
comet_chart.show()
```
