## ECH Cipher Suites over Time


```python
from pathlib import Path

import altair as alt
import polars as pl

from src.utils.date import monthYear

DATA_DIR = Path("../../data/measurements")

df = pl.read_parquet(DATA_DIR / "ech_kem.parquet")
```

```python
# Create the base chart
base = alt.Chart(df)

# Line and point layer (keep points connected)
line_chart = (
    base.mark_line(point=True)
    .encode(
        x=alt.X(
            "test_date:T",
            # title="Date"
            title=None,
            axis=alt.Axis(format=monthYear),
        ),
        y=alt.Y(
            "kem_count:Q", title="Number of Advertised KEMs", scale=alt.Scale(type="log"), axis=alt.Axis(grid=False)
        ),
    )
    .configure_legend(
        orient="bottom",
    )
    .properties(
        height=200,
    )
)

# Display the chart
line_chart.save("../../plots/ech_kem.pdf")
line_chart.show()
```
